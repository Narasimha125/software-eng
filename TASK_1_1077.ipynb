```python
import numpy as np
import matplotlib.pyplot as plt

# Hardcoded weather variables
a = 1.2  # Quadratic coefficient
b = -3.4  # Linear coefficient
c = 2.1  # Constant term
print("Narasimha  BU22CSEN0101077")
# Generating input data (e.g., time in hours)
x = np.linspace(0, 10, 100)  # 0 to 10 hours
y = a * x**2 + b * x + c  # Quadratic equation

# Plot the data
plt.plot(x, y, label="Weather Prediction")
plt.title("Weather Model (Hardcoded Variables)")
plt.xlabel("Time (hours)")
plt.ylabel("Weather Parameter")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
print("Narasimha  BU22CSEN0101077")


try:
    with open("input_single.txt", "r") as file:
        data = [line.strip() for line in file.readlines() if line.strip()]
        if len(data) < 3:
            raise ValueError("File must contain exactly three non-empty lines for coefficients.")
        a, b, c = map(float, data)


    x = np.linspace(0, 10, 100)
    y = a * x**2 + b * x + c


    plt.plot(x, y, label=f"y = {a}x + {b}x + {c}")
    plt.title("Weather Model (Single Set of Variables from File)")
    plt.xlabel("Time (hours)")
    plt.ylabel("Weather Parameter")
    plt.legend()
    plt.grid()
    plt.show()

except FileNotFoundError:
    print("Error: The file 'input_single.txt' was not found.")
except ValueError as e:
    print(f"Error: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
print("Narasimha  BU22CSEN0101077")

# Reading multiple sets of coefficients from a file
try:
    with open("input_multiple.txt", "r") as file:
        datasets = []
        for line in file:
            coefficients = list(map(float, line.split()))
            if len(coefficients) != 3:
                raise ValueError("Each line must contain exactly three numbers (a, b, c).")
            datasets.append(coefficients)

    # Plot graphs for each set of coefficients
    x = np.linspace(0, 10, 100)  # Common x-axis for all sets
    for i, (a, b, c) in enumerate(datasets):
        y = a * x**2 + b * x + c
        plt.plot(x, y, label=f"Set {i+1}: y = {a}x + {b}x + {c}")

    plt.title("Weather Model (Multiple Sets of Variables from File)")
    plt.xlabel("Time (hours)")
    plt.ylabel("Weather Parameter")
    plt.legend()
    plt.grid()
    plt.show()

except FileNotFoundError:
    print("Error: The file 'input_multiple.txt' was not found.")
except ValueError as e:
    print(f"Error: {e}")
```
