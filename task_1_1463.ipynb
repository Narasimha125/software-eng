```python
import matplotlib.pyplot as plt

# Stage 1: Hardcoded Variables
def quadratic_hardcoded():
    a, b, c, x = 2, -3, 1, 5
    y = a * x**2 + b * x + c
    print(f"Hardcoded Result: y = {y}")
    plot_graph(a, b, c)

# Plotting Function
def plot_graph(a, b, c):
    x_values = [i for i in range(-10, 11)]
    y_values = [a * x**2 + b * x + c for x in x_values]
    plt.plot(x_values, y_values, label=f"y = {a}x² + {b}x + {c}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()

# Run the first stage
quadratic_hardcoded()
```

```python
import matplotlib.pyplot as plt

# Stage 2: Keyboard Input
def quadratic_input():
    a = float(input("Enter value for a: "))
    b = float(input("Enter value for b: "))
    c = float(input("Enter value for c: "))
    x = float(input("Enter value for x: "))
    y = a * x**2 + b * x + c
    print(f"Result: y = {y}")
    plot_graph(a, b, c)

# Plotting Function
def plot_graph(a, b, c):
    x_values = [i for i in range(-10, 11)]
    y_values = [a * x**2 + b * x + c for x in x_values]
    plt.plot(x_values, y_values, label=f"y = {a}x² + {b}x + {c}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()

# Run the second stage
quadratic_input()
```

```python
import matplotlib.pyplot as plt

# Stage 3: Read from File
def quadratic_from_file(filename):
    with open(filename, 'r') as file:
        a, b, c, x = map(float, file.readline().split())
        y = a * x**2 + b * x + c
        print(f"From file: For a={a}, b={b}, c={c}, x={x}, y = {y}")
        plot_graph(a, b, c)

# Plotting Function
def plot_graph(a, b, c):
    x_values = [i for i in range(-10, 11)]
    y_values = [a * x**2 + b * x + c for x in x_values]
    plt.plot(x_values, y_values, label=f"y = {a}x² + {b}x + {c}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()

# Run the third stage
quadratic_from_file('input_single.txt')
```

```python
import matplotlib.pyplot as plt

# Stage 4: Read from File (Multiple Sets)
def quadratic_from_file_multiple(filename):
    with open(filename, 'r') as file:
        for line in file:
            a, b, c, x = map(float, line.split())
            y = a * x**2 + b * x + c
            print(f"For a={a}, b={b}, c={c}, x={x}: y = {y}")
            plot_graph(a, b, c)

# Plotting Function
def plot_graph(a, b, c):
    x_values = [i for i in range(-10, 11)]
    y_values = [a * x**2 + b * x + c for x in x_values]
    plt.plot(x_values, y_values, label=f"y = {a}x² + {b}x + {c}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()

# Run the fourth stage
quadratic_from_file_multiple('input_multiple.txt')
```
